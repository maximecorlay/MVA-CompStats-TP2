```python
import numpy as np
```

```python
N=10000
p=[0.1,0.6,0.2,0.05,0.05]

def minimum(p,u):
  if u<=p[0]:
    return 1
  elif u<=p[0]+p[1]:
    return 2
  elif u<=p[0]+p[1]+p[2]:
    return 3
  elif u<=p[0]+p[1]+p[2]+p[3]:
    return 4
  return 5

U=np.random.uniform(0,1,N)
X=np.zeros(N)
for k in range(N):
  X[k]=minimum(p,U[k])
freq=np.histogram(X,bins=5)[0]/N
print(freq)
```
